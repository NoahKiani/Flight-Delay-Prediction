```python

import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df = pd.read_csv('flights_sample_3m.csv')
df.sample(10)
```

# Data Cleaning

```python
df_clean1 = df[df['CANCELLED'] != 1] #We can't analyse flight delays when there is no flight
```

```python
print(df_clean1.isnull().sum())
```

```python
df_clean2 = df_clean1.dropna(subset=['ARR_DELAY']) #We can't work with arrival delays when the target is a NaN.
```

```python
print(df_clean2.isnull().sum())
```

We can see that with the deletion of the missing arrival delays and cancellations, all other NaN, except the delay components and cancellation code, are also gone. So we only have to deal with those ones. I chose to put zero into the delay components that contain a NaN, as I think that a NaN means this delay component does not have impact on the arrival delay that flight. I chose to do nothing with cancellation code because I ain't going to work with this feature anyway.

```python
import warnings
warnings.simplefilter(action='ignore', category=pd.errors.SettingWithCopyWarning)
df_clean2[['DELAY_DUE_CARRIER', 'DELAY_DUE_WEATHER', 'DELAY_DUE_NAS', 'DELAY_DUE_SECURITY', 'DELAY_DUE_LATE_AIRCRAFT']] = df_clean2[['DELAY_DUE_CARRIER', 'DELAY_DUE_WEATHER', 'DELAY_DUE_NAS', 'DELAY_DUE_SECURITY', 'DELAY_DUE_LATE_AIRCRAFT']].fillna(0.0)

```

```python
print(df_clean2.isnull().sum())
```

```python
df_clean2.to_csv('clean_data.csv', index=False)
df_clean2.sample(10)
```

# Delay distribution over months

This can show me if the time of the year has an impact on arrival delays. 

```python
df_clean2['FL_DATE'] = pd.to_datetime(df_clean2['FL_DATE'], errors='coerce')

df_clean2['Month'] = df_clean2['FL_DATE'].dt.month
avg_monthly_delay = df_clean2.groupby('Month')['ARR_DELAY'].mean()

plt.figure(figsize=(10, 6))
avg_monthly_delay.plot(kind='line', marker='o', color='green')
plt.title('Gemiddelde Aankomstvertraging per Maand')
plt.xlabel('Maand')
plt.ylabel('Gemiddelde Aankomstvertraging (minuten)')
plt.grid(True)
plt.show()
```

This graph says that during the vacation period in the summer and during the Holidays, the average delays are 10 times as high than other months of the year. This suggest a significant correlation between month and arrival delay.

# Delay per origin-destination (connection)

To make a heatmap of all origin airports would be too unclear, because there are hundreds of airports. So I filtered the busiest airports to see if these have any correlation to each other.

```python
top_origins = df_clean2['ORIGIN'].value_counts().nlargest(10).index
top_destinations = df_clean2['DEST'].value_counts().nlargest(10).index


filtered_data = df_clean2[df_clean2['ORIGIN'].isin(top_origins) & df_clean2['DEST'].isin(top_destinations)]


delay_matrix = filtered_data.pivot_table(index='ORIGIN', columns='DEST', values='ARR_DELAY', aggfunc='mean')


plt.figure(figsize=(10, 6))
sns.heatmap(delay_matrix, cmap='coolwarm', annot=True)
plt.title('Gemiddelde Aankomstvertraging per Vertrek- en Bestemmingsluchthaven (Top 10)')
plt.xlabel('Bestemming')
plt.ylabel('Vertrekluchthaven')
plt.show()
```

Out of this graph I can conclude that certain routes do have different average arrival delays. The connection DFW (Dallas) - MCO (Orlando) has 14 minutes average delays while other connections have negative average delays. Origin-Destination could be interesting to look at when doing EDA and model training.

# Percentage per delay category

To prove the point being made about filtering the data, this graph will be made.

```python
def categorize_delay(arr_delay):
    if arr_delay < 0:
        return 'Op tijd'
    elif arr_delay <= 10:
        return 'Tot 10 minuten'
    elif arr_delay <= 30:
        return 'Tot 30 minuten'
    elif arr_delay <= 60:
        return 'Tot 60 minuten'
    else:
        return 'Meer dan 60 minuten'

# Nieuwe kolom met vertraging categorieën toevoegen
df_clean2['DELAY_CATEGORY'] = df_clean2['ARR_DELAY'].apply(categorize_delay)

# Aantal vluchten per categorie
delay_counts = df_clean2['DELAY_CATEGORY'].value_counts()

# Percentages berekenen
delay_percentages = (delay_counts / delay_counts.sum()) * 100

# Visualisatie - Staafdiagram
plt.figure(figsize=(10, 6))
delay_percentages.plot(kind='bar', color='skyblue')
plt.title('Percentage van Vluchten per Vertraging Categorie')
plt.xlabel('Vertraging Categorie')
plt.ylabel('Percentage van Vluchten (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Voeg percentage labels toe boven de staven
for i, v in enumerate(delay_percentages):
    plt.text(i, v + 1, f"{v:.2f}%", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

This graph shows indeed that only 6% of all flights are delayed more than 60 minutes. When working with this 6%, things are a bit more clear. When machine teaching, all the data should of course be included. 

# Delayed flights per distance category

To make another graph for the distance, I chose a bar plot. This plot shows the percentages of flights delayed more than 30 minutes. I chose 30 minutes because I think this is the point that discomfort with passengers will occur.

```python
max_distance = 5812


bins = [0, 1000, 2000, 3000, 4000, 5000, 6000]
labels = ['0-1000', '1000-2000', '2000-3000', '3000-4000', '4000-5000', '5000-6000']


df_clean2['DISTANCE_CATEGORY'] = pd.cut(df_clean2['DISTANCE'], bins=bins, labels=labels, right=False)
df_clean2.to_csv('schoon.csv', index=False) #this csv file will be used in other notebooks

delay_counts = df_clean2[df_clean2['ARR_DELAY'] > 30]['DISTANCE_CATEGORY'].value_counts()
total_counts = df_clean2['DISTANCE_CATEGORY'].value_counts()


delay_percentages = (delay_counts / total_counts) * 100


plt.figure(figsize=(10, 10))
delay_percentages.plot(kind='bar', color='lightcoral')
plt.title('Percentage van Vluchten met Meer dan 30 Minuten Vertraging per Afstand')
plt.xlabel('Afstand (mijlen)')
plt.ylabel('Percentage van Vluchten met >30 Minuten Vertraging (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')


for i, v in enumerate(delay_percentages):
    plt.text(i, v - 2, f"{v:.1f}%", ha='center', va='top', color='black')

plt.tight_layout()
plt.show()
```

This graph shows that every distance categories are relatively equal in delays above 30 minutes. The next step is to dig deeper into EDA.
